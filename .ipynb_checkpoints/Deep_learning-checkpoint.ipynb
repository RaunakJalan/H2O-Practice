```python
import h2o
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h2o.init()
```

```python
data = h2o.import_file("http://s3.amazonaws.com/h2o-public-test-data/smalldata/airlines/allyears2k_headers.zip")
```

```python
train, valid, test = data.split_frame([0.8,0.1], seed = 42)
```

```python
print("%d/%d/%d" % (train.nrows, valid.nrows, test.nrows))
```

```python
y = "IsArrDelayed"
ignoreFields = [
    'ArrDelay',
    'DepDelay','CarrierDelay',
    'WeatherDelay',
    'NASDelay',
    'SecurityDelay',
    'LateAircraftDelay','IsArrDelayed',
    'IsDepDelayed','ActualElapsedTime'
]
xAll = [i for i in train.names if i not in ignoreFields]
```

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
```

```python
m_def = H2ODeepLearningEstimator()
%time m_def.train(xAll, y, train, validation_frame=valid)
```

```python
m_def.model_performance(test)
```

```python
m_def
```

```python
m_def.plot()
```

### Tuning

```python
m_200_epochs = H2ODeepLearningEstimator(epochs=200,
                                       stopping_rounds=5,
                                       stopping_tolerance=0,
                                       stopping_metric="logloss"
                                       )
#Above all are defaults

%time m_200_epochs.train(xAll,y,train,validation_frame=valid)
```

```python
m_200_epochs.model_performance(test)
```

```python
m_200_epochs.plot()
```

```python
m_200x200x200_epochs = H2ODeepLearningEstimator(epochs=200,
                                       stopping_rounds=5,
                                       stopping_tolerance=0,
                                       stopping_metric="logloss",
                                       hidden = [200,200,200]
                                       )
%time m_200x200x200_epochs.train(xAll,y,train,validation_frame=valid)
```

```python
m_200x200x200_epochs.model_performance(test)
```

```python
m_200x200x200_epochs.plot()
```
